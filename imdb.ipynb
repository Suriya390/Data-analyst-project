# Import the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('imdb_top_1000.csv')
df
```

# Data Cleaning

Checking null values

```python
df.isnull().sum()
```

```python
df['Certificate']=df['Certificate'].fillna('Unrated')
```

```python
df['Meta_score']=df['Meta_score'].fillna(df['Meta_score'].mean())
```

```python
df['Gross']=df['Gross'].fillna(0)
```

```python
df[['Certificate', 'Meta_score', 'Gross']].isnull().sum()
```

removing special characters

```python
df['Runtime'] = pd.to_numeric(df['Runtime'].astype(str).str.replace(' min', '', regex=False))
df['Gross'] = df['Gross'].astype(str).str.replace(',', '').str.strip()
df['Gross'] = pd.to_numeric(df['Gross'], errors='coerce')
 
```

Check and remove duplicate values

```python
df.duplicated().any()
df.drop_duplicates(inplace=True)
```

changing column datatype

```python
df=df.astype({'Gross':int, 'Runtime':int})
```

```python
df = df[df['Released_Year'].notna()]  # Removes NaNs
df = df[df['Released_Year'].astype(str).str.isnumeric()]
```

# Data Transformation

Changing columns

```python
df['Genre'] = df['Genre'].str.lower().str.strip()
df['Certificate'] = df['Certificate'].str.upper().str.strip()
```

Removing unwanted columns

```python
df=df.drop(['Poster_Link', 'Star1', 'Star2', 'Star3', 'Star4', 'Director'], axis=1)
```

creating new column for analysis

```python
df['Rating_Level'] = pd.cut(df['IMDB_Rating'], bins=[0, 6, 7.5, 9, 10], labels=['Low', 'Medium', 'High', 'Top'])
```

```python
df
```

Sorting index

```python
df=df.reset_index(drop=True)
```

# Plot Analysis

1. What is the distribution of IMDb ratings across all movies?

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['IMDB_Rating'], bins=20, color='skyblue', edgecolor='black')

# Customize the plot
plt.title('Distribution of IMDb Ratings')
plt.xlabel('IMDb Rating')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()
```

# Observation

* The most movies rating are there in 7.75 to 8
* The second ratings are in 8.25 to 8.50
* Lowest ratings are in 9 and 9.25

2. How do IMDb ratings vary across different genres?

```python
top_genres = df['Genre'].value_counts().nlargest(5).index
df_top_genres = df[df['Genre'].isin(top_genres)]


plt.figure(figsize=(10, 4))
sns.boxplot(data=df_top_genres, x='Genre', y='IMDB_Rating')
plt.title('IMDb Ratings by Genre (Top 5 Genres)')
plt.xlabel('Genre')
plt.ylabel('IMDb Rating')
plt.tight_layout()
plt.show()
```

# Observation

* The more rating are in drama genre
* Second high ratings are in actionn, crime, drama
* Drama and romance are near to rating 8 
* The lowest ratings are in comedy, drama, romance and comedy, drama

3. What are the most common genres

```python
plt.figure(figsize=(25, 15))
sns.countplot(data=df, x='Genre', order=df['Genre'].value_counts().index)
plt.title('Most Common Movie Genres')
plt.xticks(rotation=90)
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()
```

# Observation

* The highest movie count in dram genre
* The second movie count in drama romace, comedy drama, comedy drama romance, action crime drama, and biography drama history

4. How has the average IMDb rating changed over the years?

```python
df['Released_Year'] = df['Released_Year'].astype(int)
yearly_avg = df.groupby('Released_Year')['IMDB_Rating'].mean().reset_index()


plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_avg, x='Released_Year', y='IMDB_Rating', marker='o')
plt.title('Average IMDb Rating Over the Years')
plt.xlabel('Year')
plt.ylabel('Average IMDb Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Observation

* The average ratings are in 1920 to 1940
* 1940 to 2000 are mid range ratings happened 7.8 to 8.3
* 2000 to 2020 are lowest range ratings because the ratings are in 7.8 to 8

5. How has the number of top movies per year changed over time?

```python
movies_per_year = df['Released_Year'].value_counts().sort_index()

plt.figure(figsize=(14, 6))
sns.lineplot(x=movies_per_year.index, y=movies_per_year.values)
plt.title('Number of Top Movies per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Observation

* There is less top movies between 1920 to 1940 but the top movies slightly increases in 1940 to 1960
* 1960 to 1980 increases slightly high
* The top highest movies are in 1980 to 2020

6. Finding correlation between Run time and IMDB rating

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Runtime', y='IMDB_Rating')
plt.title('Correlation Between Runtime and IMDb Rating')
plt.xlabel('Runtime (minutes)')
plt.ylabel('IMDb Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Observation

* There is a positive correlation between runtime and IMDB rating
* The most movie raings fall in 50 to 150 minutes
* The high rated movies are in 8.0 to 8.5 and the runtime is 100 to 200

7. Which runtime range has the highest concentration of top-rated movies?

```python
bins = [0, 6, 7, 8, 9, 10]
labels = ['<6', '6-7', '7-8', '8-9', '9-10']
df['Rating_Bucket'] = pd.cut(df['IMDB_Rating'], bins=bins, labels=labels, include_lowest=True)

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Rating_Bucket', y='Runtime')
plt.title('Movie Runtime Distribution by IMDb Rating')
plt.xlabel('IMDb Rating')
plt.ylabel('Runtime (minutes)')
plt.tight_layout()
plt.show()
```

# Observation

* There is a correlation between runtime and rating
* The high rated movies 8 - 9 and 9 - 10 are runtime in 150 to 175 minutes

8. How does the number of votes relate to IMDb rating?

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='No_of_Votes', y='IMDB_Rating', data=df, alpha=0.5)

plt.xscale('log')
plt.title('Relationship Between Number of Votes and IMDb Rating')
plt.xlabel('Number of Votes (log scale)')
plt.ylabel('IMDb Rating')
plt.tight_layout()
plt.show()
```

# Observation

* There is a correlation between Rating and votes
* The more movie ratings fall in 7.5 - 8.5
* The high rated movies 8.5 - 9.25 has the highest number of votes

9. How do rating, votes, and gross earnings interact together?

```python
df = df[(df['No_of_Votes'] > 0) & (df['Gross'] > 0)]

# Plot
plt.figure(figsize=(12, 7))
sns.scatterplot(data=df, x='No_of_Votes', y='IMDB_Rating', hue='IMDB_Rating', size='Gross', sizes=(20, 300), palette='viridis', alpha=0.6,  legend='brief')
plt.xscale('log')
plt.title('Interaction Between IMDb Rating, Votes, and Gross Earnings')
plt.xlabel('Number of Votes (log scale)')
plt.ylabel('IMDb Rating')
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

# Observation

* As the number of votes increases the IMDB rating also increases it shows the relation
* The more movies rating are in 7.5 and 8.5
* The highest rated movies 9.0 have a large number of votes
