# Data cleaning

Renaming the column

```python
df=df.rename(columns={'Name of State / UT':'State_Name','Total Confirmed cases':'Total_confirmed_cases','Cured/Discharged/Migrated':'Recovered','New cases':'New_cases','New deaths':'New_deaths','New recovered':'New_recovered'})
```

Datatype conversion

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df
```

```python
df.astype({'Total_confirmed_cases':int,'Recovered':int})
```

Finding null values

```python
df.duplicated().any()
```

Handling Missing values

```python
for col in ['New_cases','New_deaths','New_recovered']:
    df[col]=df[col].fillna(0).astype(int)
```

Handling NaN values

```python
for col in ['Total_confirmed_cases', 'Death', 'Recovered']:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)
```

# Data Transformation

Sorting data by date and state name

```python
df.sort_values(by=['State_Name', 'Date'], inplace=True)
```

Removing unwanted columns

```python
df=df.drop(['Latitude','Longitude'],axis=1)
```

```python
df.reset_index(drop=True)
```

Calculate active cases

```python
df['Active_cases'] = df['Total_confirmed_cases'] - df['Death'] - df['Recovered']
df
```

Calculate recovery rate

```python
df['Recovery_Rate'] = (df['Recovered'] / df['Total_confirmed_cases']).fillna(0) * 100
df['Recovery_Rate']=df['Recovery_Rate'].round(2)
```

Rearranging columns

```python
lst=['Date','State_Name','Total_confirmed_cases','Death','Recovered','Active_cases','New_cases','New_deaths','New_recovered','Recovery_Rate']
df=df[lst]
```

Reset index numbers

```python
df=df.reset_index(drop=True)
```

# Plot analysis

```python
df.columns
```

1. How did the total confirmed COVID-19 cases change over in time?

```python
daily_confirmed = df.groupby('Date')['Total_confirmed_cases'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=daily_confirmed, x='Date', y='Total_confirmed_cases')
plt.title('Total Confirmed COVID-19 Cases Over Time in India')
plt.xlabel('Date')
plt.ylabel('Total Confirmed Cases')
plt.tight_layout()
```

# observation

* The total confirmed cases are low but it increase month by month
* In the month of may it  started to increases
* In the August month total confirmed case reached the 2 million

2. What is the trend of death overtime

```python
daily_deaths=df.groupby('Date')['Death'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=daily_deaths, x='Date', y='Death')
plt.title('Total COVID-19 Deaths Over Time in India')
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.tight_layout()
plt.show()
```

# Observation

* At the start no of deaths are nearly to 0
* The death increases at May
* In August month the death reaches 40,000

3. Which states had the highest recovery growth over time?

```python
New_data = df[df['Date'] == df['Date'].max()]
top_states = New_data.groupby('State_Name')['Total_confirmed_cases'].sum().nlargest(5).index.tolist()
df_top_recovery = df[df['State_Name'].isin(top_states)]
df_top_recovery = df_top_recovery.sort_values(['State_Name', 'Date'])

plt.figure(figsize=(14, 7))
sns.lineplot(data=df_top_recovery, x='Date', y='Total_confirmed_cases', hue='State_Name')
plt.title('Recovery Growth Over Time (Top 5 States)')
plt.xlabel('Date')
plt.ylabel('Cumulative Recoveries')
plt.xticks(rotation=45)
plt.legend(title='State/UT')
```

# Observation

* In the top 5 recovery states Maharashtra holds the first place because it recovers 4,50,000 cases in August
* The second highest is the Tamil Nadu it recovers more than 2,00,000
* Other 3 states Andhra Pradhesh, Delhi, and Karnataka are recovers more than 1,00,000 cases
* The most recovers are happened only in august month

4. Which State have the highest confirmed cases

```python
state=df.groupby('State_Name')['Total_confirmed_cases'].max().sort_values(ascending=False)

plt.figure(figsize=(12,7))
c=sns.barplot(x=state.values, y=state.index)
c.bar_label(c.containers[0])
plt.title('Maximum Total Confirmed COVID-19 Cases per State')
plt.xlabel('Total Confirmed Cases')
plt.ylabel('State / UT')
plt.tight_layout()
plt.show()
```

# Observation

* The most highest confirmed cases are in Maharashtra the total cases are 4,68,265
* The second highest cases are Tamil Nadu(2,73,460) and Andhra Pradesh(1,86,461)
* The mid range confirmed case states are Karnataka, Delhi, and Uttar Pradesh because more than 1,00,00 confirmed cases are there
* The other states are lowest confirmed cases because these all are lower than 85,000 cases

6. Relationship between confirmed cases and deaths

```python
latest_date = df['Date'].max()
latest_data = df[df['Date'] == latest_date]

plt.figure(figsize=(12,6))
sns.scatterplot(data=latest_data, x='Total_confirmed_cases',y='Death', hue='State_Name', palette='Set2', edgecolor='black')
plt.title('Relationship Between Total Confirmed Cases and Deaths (per State)', fontsize=16)
plt.xlabel('Total Confirmed Cases')
plt.ylabel('Total Deaths')
plt.tight_layout()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='State')
plt.show()
```

# Observation

* The total confirmed cases increases total death also increases
* The outlier is Maharashtra because the highest confirmed cases and the highest death are there in Maharashtra

8. Top 5 states confirmed cases
